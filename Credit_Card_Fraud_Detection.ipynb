Visualizing Class Imbalance

```python
df['Class'].value_counts().plot(kind='bar')

plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

Handling Class imbalances

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Split into features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

After handling class imbalanaces using SMOTE

```python
pd.Series(y_train_sm).value_counts().plot(kind='bar', color=['skyblue', 'salmon'])

plt.title("Class Distribution After SMOTE")
plt.xticks([0, 1], ['Not Fraud (0)', 'Fraud (1)'], rotation=0)
plt.ylabel("Count")
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

Intialising and training the model


```python
rf = RandomForestClassifier(
    n_estimators=100,      # fewer trees (for now)
    max_depth=15,         # limit tree depth
    n_jobs=-1,            # use all CPU cores
    random_state=42
)
rf.fit(X_train_sm, y_train_sm)
```

```python
X_train_sm.shape
```

Testing the model

```python
ypred = rf.predict(X_test)
```

Accuracy Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,ypred))
```

```python
confusion_matrix(y_test,ypred)
```
